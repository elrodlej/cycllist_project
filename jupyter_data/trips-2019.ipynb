```python
import pandas as pd

trips_2019_Q1 = pd.read_csv('Trips_2019_Q1.csv')
trips_2019_Q2 = pd.read_csv('Trips_2019_Q2.csv')
trips_2019_Q3 = pd.read_csv('Trips_2019_Q3.csv')
trips_2019_Q4 = pd.read_csv('Trips_2019_Q4.csv')
```

In the case of this year compared to the previous ones, the format changed for Q1, so following I will edit the column names, date formats and data type of trip duration

```python
trips_2019_Q2.head(5)
```

```python
wrong_names = list(trips_2019_Q2.columns)
print(wrong_names)
```

```python
#I get the names I want to assign to the Q1 table.
correct_names = list(trips_2019_Q1.columns)
```

```python
#Here I replace the column names with the names I want using a dictionary for it
name_replace = dict(zip(wrong_names, correct_names))
```

```python
trips_2019_Q2.rename(columns=name_replace, inplace=True)
```

```python
trips_2019_Q2.head(5)
```

Now I will check if the format of all the tables matches

```python
trips_2019_Q1.head(5)
```

It does, but some data does not match with previous years, so I will change the format after concatenating the data here.

```python
trips_2019 = pd.concat([trips_2019_Q1, trips_2019_Q2, trips_2019_Q3, trips_2019_Q4])
```

```python
trips_2019.head(5)
```

```python
trips_2019['birthyear'] = trips_2019['birthyear'].fillna(-1).astype(int)
```

```python
trips_2019.sample(20)
```

```python
trips_2019['tripduration'] = trips_2019['tripduration'].str.replace(',','').astype(float).astype(int)
```

```python
trips_2019.sample(5)
```

Now I have to change date times to match previous tables

```python
#I will change the date format to match what we had in previous years
trips_2019['start_time'] = pd.to_datetime(trips_2019['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2019['end_time'] = pd.to_datetime(trips_2019['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

```python
trips_2019.sample(10)
```

Now I will save the new table as csv, this will be openned and processed using SQL 
in MySQL.

```python
trips_2019.to_csv('trips_2019.csv', index=False)
```
