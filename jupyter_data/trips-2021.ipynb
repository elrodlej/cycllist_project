```python
import pandas as pd

trips_2021_jan = pd.read_csv('trip_2021_jan.csv')
trips_2021_feb = pd.read_csv('trip_2021_feb.csv')
trips_2021_mar = pd.read_csv('trip_2021_mar.csv')
trips_2021_apr = pd.read_csv('trip_2021_apr.csv')
trips_2021_may = pd.read_csv('trip_2021_may.csv')
trips_2021_jun = pd.read_csv('trip_2021_jun.csv')
trips_2021_jul = pd.read_csv('trip_2021_jul.csv')
trips_2021_aug = pd.read_csv('trip_2021_aug.csv')
trips_2021_sep = pd.read_csv('trip_2021_sep.csv')
trips_2021_oct = pd.read_csv('trip_2021_oct.csv')
trips_2021_nov = pd.read_csv('trip_2021_nov.csv')
trips_2021_dec = pd.read_csv('trip_2021_dec.csv')
```

```python
trips_2021_dec.columns
```

I will concat these tables and then I will change some of the format to fit previous years

```python
trips_2021 = pd.concat([trips_2021_jan, trips_2021_feb, trips_2021_mar, trips_2021_apr, trips_2021_may, trips_2021_jun, trips_2021_jul, trips_2021_aug, trips_2021_sep, trips_2021_oct, trips_2021_nov, trips_2021_dec])
```

```python
trips_2021.sample(10)
```

```python
old_names = ['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']
new_names = ['trip_id', 'start_time', 'end_time', 'from_station_name', 'from_station_id', 'to_station_name', 'to_station_id', 'usertype']

change_names = dict(zip(old_names, new_names))
```

```python
trips_2021.rename(columns=(change_names), inplace=True)
```

```python
trips_2021.head()
```

Now I will start cleaning the start_time, end time, I will have to create the column tripduration in this case too

```python
trips_2021['start_time'] = pd.to_datetime(trips_2021['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2021['end_time'] = pd.to_datetime(trips_2021['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

```python
trips_2021.head(5)
```

Next I will get the tripduration column

```python
trips_2021['start_time'] = pd.to_datetime(trips_2021['start_time'], format='%m/%d/%Y %H:%M:%S')
trips_2021['end_time'] = pd.to_datetime(trips_2021['end_time'], format='%m/%d/%Y %H:%M:%S')
```

```python
trips_2021['tripduration'] = (trips_2021['end_time'] - trips_2021['start_time']).dt.total_seconds()
```

```python
trips_2021.head()
```

```python
trips_2021['tripduration'] = trips_2021['tripduration'].astype(int)
```

```python
trips_2021.head()
```

As with 2020, I will check if there are null values in the lat columns

```python
print(trips_2021['start_lat'].isnull().any())
```

```python
print(trips_2021['end_lat'].isnull().any())
```

```python
num_nulls = trips_2021['end_lat'].isnull().sum()
print(num_nulls)
```

```python
trips_2021 = trips_2021.dropna(subset= ['end_lat'])
```

```python
print(trips_2021['end_lat'].isnull().any())
```

Now I will format the date columns once more

```python
trips_2021['start_time'] = pd.to_datetime(trips_2021['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2021['end_time'] = pd.to_datetime(trips_2021['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

Now that I have all I need, I will save this dataframe

```python
trips_2021.to_csv('trips_2021.csv', index= False)
```
