```python
import pandas as pd

trips_2022_jan = pd.read_csv('trip_2022_jan.csv')
trips_2022_feb = pd.read_csv('trip_2022_feb.csv')
trips_2022_mar = pd.read_csv('trip_2022_mar.csv')
trips_2022_apr = pd.read_csv('trip_2022_apr.csv')
trips_2022_may = pd.read_csv('trip_2022_may.csv')
trips_2022_jun = pd.read_csv('trip_2022_jun.csv')
trips_2022_jul = pd.read_csv('trip_2022_jul.csv')
trips_2022_aug = pd.read_csv('trip_2022_aug.csv')
trips_2022_sep = pd.read_csv('trip_2022_sep.csv')
trips_2022_oct = pd.read_csv('trip_2022_oct.csv')
trips_2022_nov = pd.read_csv('trip_2022_nov.csv')
trips_2022_dec = pd.read_csv('trip_2022_dec.csv')
```

```python
trips_2022_jan.head()
```

I will concat these tables and then I will change some of the format to fit previous years

```python
trips_2022 = pd.concat([trips_2022_jan, trips_2022_feb, trips_2022_mar, trips_2022_apr, trips_2022_may, trips_2022_jun, trips_2022_jul, trips_2022_aug, trips_2022_sep, trips_2022_oct, trips_2022_nov, trips_2022_dec])
```

```python
trips_2022.sample(10)
```

```python
old_names = ['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']
new_names = ['trip_id', 'start_time', 'end_time', 'from_station_name', 'from_station_id', 'to_station_name', 'to_station_id', 'usertype']

change_names = dict(zip(old_names, new_names))
```

```python
trips_2022.rename(columns=(change_names), inplace=True)
```

```python
trips_2022.head()
```

Now I will start cleaning the start_time, end time, I will have to create the column tripduration in this case too

```python
trips_2022['start_time'] = pd.to_datetime(trips_2022['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2022['end_time'] = pd.to_datetime(trips_2022['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

```python
trips_2022.head(5)
```

Next I will get the tripduration column

```python
trips_2022['start_time'] = pd.to_datetime(trips_2022['start_time'], format='%m/%d/%Y %H:%M:%S')
trips_2022['end_time'] = pd.to_datetime(trips_2022['end_time'], format='%m/%d/%Y %H:%M:%S')
```

```python
trips_2022['tripduration'] = (trips_2022['end_time'] - trips_2022['start_time']).dt.total_seconds()
```

```python
trips_2022.head()
```

```python
trips_2022['tripduration'] = trips_2022['tripduration'].astype(int)
```

```python
trips_2022.head()
```

As before, I will check if there are null values in the lat columns

```python
print(trips_2022['start_lat'].isnull().any())
```

```python
print(trips_2022['end_lat'].isnull().any())
```

```python
print(trips_2022['end_lat'].isnull().sum())
```

```python
trips_2022 = trips_2022.dropna(subset=['end_lat'])
```

Now I will change the dates columns format

```python
trips_2022['start_time'] = pd.to_datetime(trips_2022['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2022['end_time'] = pd.to_datetime(trips_2022['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

Now that I have all I need, I will save this dataframe

```python
trips_2022.to_csv('trips_2022.csv', index= False)
```
