```python
import pandas as pd

trips_2020_Q1 = pd.read_csv('Trips_2020_Q1.csv')
trips_2020_apr = pd.read_csv('trip_2020_apr.csv')
trips_2020_may = pd.read_csv('trip_2020_may.csv')
trips_2020_jun = pd.read_csv('trip_2020_jun.csv')
trips_2020_jul = pd.read_csv('trip_2020_jul.csv')
trips_2020_aug = pd.read_csv('trip_2020_aug.csv')
trips_2020_sep = pd.read_csv('trip_2020_sep.csv')
trips_2020_oct = pd.read_csv('trip_2020_oct.csv')
trips_2020_nov = pd.read_csv('trip_2020_nov.csv')
trips_2020_dec = pd.read_csv('trip_2020_dec.csv')
```

This year has a bigger change, data is now stored monthly, so the code used for data cleaning is different

```python
trips_2020_dec.head(5)
```

I will concat these tables and then I will change some of the format to fit previous years

```python
trips_2020 = pd.concat([trips_2020_Q1, trips_2020_apr, trips_2020_may, trips_2020_jun, trips_2020_jul, trips_2020_aug, trips_2020_sep, trips_2020_oct, trips_2020_nov, trips_2020_dec])
```

```python
trips_2020.sample(10)
```

```python
old_names = ['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']
new_names = ['trip_id', 'start_time', 'end_time', 'from_station_name', 'from_station_id', 'to_station_name', 'to_station_id', 'usertype']

change_names = dict(zip(old_names, new_names))
```

```python
trips_2020.rename(columns=(change_names), inplace=True)
```

```python
trips_2020.tail()
```

Now I will start cleaning the start_time, end time, I will have to create the column tripduration in this case too

```python
trips_2020['start_time'] = pd.to_datetime(trips_2020['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2020['end_time'] = pd.to_datetime(trips_2020['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

```python
trips_2020.head(5)
```

Next I will get the tripduration column

```python
trips_2020['start_time'] = pd.to_datetime(trips_2020['start_time'], format='%m/%d/%Y %H:%M:%S')
trips_2020['end_time'] = pd.to_datetime(trips_2020['end_time'], format='%m/%d/%Y %H:%M:%S')
```

```python
trips_2020['tripduration'] = (trips_2020['end_time'] - trips_2020['start_time']).dt.total_seconds()
```

```python
trips_2020.head()
```

```python
trips_2020['tripduration'] = trips_2020['tripduration'].astype(int)
```

I will check if there are null values in the columns regarding latitude and longitude, because they will be used for analysis.

```python
print(trips_2020['start_lat'].isnull().any())
```

```python
print(trips_2020['end_lat'].isnull().any())
```

Now I will drop rows with null values inside either of these columns

```python
trips_2020 = trips_2020.dropna(subset= ['end_lat'])
```

```python
print(trips_2020['end_lat'].isnull().any())
```

Next I will format the dates columns once more

```python
trips_2020['start_time'] = pd.to_datetime(trips_2020['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2020['end_time'] = pd.to_datetime(trips_2020['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

Now that I have all I need, I will save this dataframe

```python
trips_2020.to_csv('trips_2020.csv', index= False)
```
