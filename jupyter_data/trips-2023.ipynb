```python
import pandas as pd

trips_2023_jan = pd.read_csv('trip_2023_jan.csv')
trips_2023_feb = pd.read_csv('trip_2023_feb.csv')
trips_2023_mar = pd.read_csv('trip_2023_mar.csv')
trips_2023_apr = pd.read_csv('trip_2023_apr.csv')
```

```python
trips_2023_jan.head()
```

I will concat these tables and then I will change some of the format to fit previous years

```python
trips_2023 = pd.concat([trips_2023_jan, trips_2023_feb, trips_2023_mar, trips_2023_apr])
```

```python
trips_2023.sample(10)
```

```python
old_names = ['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']
new_names = ['trip_id', 'start_time', 'end_time', 'from_station_name', 'from_station_id', 'to_station_name', 'to_station_id', 'usertype']

change_names = dict(zip(old_names, new_names))
```

```python
trips_2023.rename(columns=(change_names), inplace=True)
```

```python
trips_2023.head()
```

Now I will start cleaning the start_time, end time, I will have to create the column tripduration in this case too

```python
trips_2023['start_time'] = pd.to_datetime(trips_2023['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2023['end_time'] = pd.to_datetime(trips_2023['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

```python
trips_2023.head(5)
```

Next I will get the tripduration column

```python
trips_2023['start_time'] = pd.to_datetime(trips_2023['start_time'], format='%m/%d/%Y %H:%M:%S')
trips_2023['end_time'] = pd.to_datetime(trips_2023['end_time'], format='%m/%d/%Y %H:%M:%S')
```

```python
trips_2023['tripduration'] = (trips_2023['end_time'] - trips_2023['start_time']).dt.total_seconds()
```

```python
trips_2023.head()
```

```python
trips_2023['tripduration'] = trips_2023['tripduration'].astype(int)
```

```python
trips_2023.head()
```

Now I will check any missing values

```python
print(trips_2023['start_lat'].isnull().any())
```

```python
print(trips_2023['end_lat'].isnull().any())
```

```python
print(trips_2023['end_lat'].isnull().sum())
```

```python
trips_2023 = trips_2023.dropna(subset=['end_lat'])
```

```python
print(trips_2023['end_lat'].isnull().any())
```

Now I will modify date columns format

```python
trips_2023['start_time'] = pd.to_datetime(trips_2023['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2023['end_time'] = pd.to_datetime(trips_2023['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

Now that I have all I need, I will save this dataframe

```python
trips_2023.to_csv('trips_2023.csv', index= False)
```
