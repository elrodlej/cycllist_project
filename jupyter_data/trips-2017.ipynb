```python
import pandas as pd

trips_2017_Q1 = pd.read_csv('Trips_2017_Q1.csv')
trips_2017_Q2 = pd.read_csv('Trips_2017_Q2.csv')
trips_2017_Q3 = pd.read_csv('Trips_2017_Q3.csv')
trips_2017_Q4 = pd.read_csv('Trips_2017_Q4.csv')
```

```python
trips_2017 = pd.concat([trips_2017_Q1, trips_2017_Q2, trips_2017_Q3, trips_2017_Q4])
```

```python
trips_2017.sample(20)
```

```python
trips_2017['birthyear'] = trips_2017['birthyear'].fillna(-1).astype(int)
```

```python
trips_2017.sample(20)
```

```python
trips_2017['start_time'] = pd.to_datetime(trips_2017['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2017['end_time'] = pd.to_datetime(trips_2017['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

```python
trips_2017.sample(10)
```
