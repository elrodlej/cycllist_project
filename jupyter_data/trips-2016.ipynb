#Now I will merge the tables so I have single tables per year, making analysis easier

```python
# I add the missing tables

apr_2016 = pd.read_csv('Trips_2016_apr.csv')
may_2016 = pd.read_csv('Trips_2016_may.csv')
jun_2016 = pd.read_csv('Trips_2016_jun.csv')
q3_2016 = pd.read_csv('Trips_2016_Q3.csv')
q4_2016 = pd.read_csv('Trips_2016_Q4.csv')
```

Then I add all of the raw tables together to see if there are no problems between them

```python
incomplete_2016 = pd.concat([apr_2016, may_2016, jun_2016, q3_2016, q4_2016])
```

```python
incomplete_2016.sample(20)
```

They were added correctly, afterwards I will clean the birthyear column with the code used 
previously.

```python
incomplete_2016['birthyear'] = incomplete_2016['birthyear'].fillna(-1).astype(int)
```

```python
incomplete_2016.sample(20)
```

```python
trips_2016 = pd.concat([trips_2016_Q1, incomplete_2016])
```

I will change the date format too, to make it easier to work with later on

```python
trips_2016.rename(columns={'starttime':'start_time', 'stoptime':'end_time'}, inplace=True)
```

```python
trips_2016['start_time'] = pd.to_datetime(trips_2016['start_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
trips_2016['end_time'] = pd.to_datetime(trips_2016['end_time']).dt.strftime('%m/%d/%Y %H:%M:%S')
```

Now I will save the new table as csv, this will be openned and processed using SQL 
in MySQL.

```python
trips_2016['birthyear'] = trips_2016['birthyear'].fillna(-1).astype(int)
```
